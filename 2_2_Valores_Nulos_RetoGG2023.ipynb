# **Limpieza de Bases de Datos**

Limpiaremos las bases de datos del socio formador para poderlo utilizar en nuestro reto a futuro.

```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargar archivo csv desde equipo
#from google.colab import files
#files.upload()
```

```python
#Carga desde un archivo de excel
df= pd.read_excel("Gastos y costos 20-23.xlsx", skiprows=5)
df
```

```python
df.info()
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

Considerando las columnas de POLIZA así como las columna de GASTO vemos que los valores que faltan son más de la mitad de los datos ingresados en la base de datos. Es mejor simplemente dar de baja las columnas de la base de datos considerando que no se tiene la informacíon necesaria para hacer una sustitución como una media o una mediana.

```python
df=df.drop(['POLIZA','GASTO'], axis=1)
df
```

```python
df.info()
```

```python
valores_nulos1=df.isnull().sum()
valores_nulos1
```

Como podemos ver tenemos ahora columnas con datos NA menores a un 25% por los que se hará una una sustitución por datos anteriores, medias o medianas.

```python
df1=df.copy()
df1
```

```python
df1['TIPO']=df1['TIPO'].fillna(method='bfill')
valores_nulos2=df1.isnull().sum()
valores_nulos2
```

```python
df1['IVA']=df1['IVA'].fillna(round(df1['TOTAL SAT']*0.1379,1))
valores_nulos3=df1.isnull().sum()
valores_nulos3
```

```python
df1['IMPORTE']=df1['IMPORTE'].fillna(round(df['IMPORTE'].median(),1))
valores_nulos4=df1.isnull().sum()
valores_nulos4
```

```python
df1['TC']=df1['TC'].fillna(1)
valores_nulos5=df1.isnull().sum()
valores_nulos5
```

```python
df1['FOLIO']=df1['FOLIO'].fillna(000000)
valaores_nulos6=df1.isnull().sum()
valaores_nulos6
```

```python
#Convertir DataFrame a CSV
df1.to_excel('Gastos y costos 20-23 (SN).xlsx')
```
